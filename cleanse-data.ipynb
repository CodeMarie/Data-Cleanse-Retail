```python
import numpy as np
import os 
import pandas as pd
```

```python
#read of branch list 
region_df = pd.read_excel('Non-branch-data/branch_list.xlsx')
region_df
```

```python
# read of products csv
productslist_df = pd.read_csv('Non-branch-data/products_list.csv')
productslist_df
```

```python
# read of branch expenses file 
branch_expenses_df = pd.read_excel('Non-branch-data/branch_expenses.xlsx')
branch_expenses_df
```

```python
# Expenses - Addition of a total expenses column ready for the profitability calculation needed 
branch_expenses_df['total_expenses'] = branch_expenses_df.apply(lambda row: row['operational_cost'] + row['staff_bonuses'] + row['misc_expenses']  + row['waste_cost'], axis=1)
branch_expenses_df
```

```python
# Creation of the preferred header terms see terminology.txt for more information 
new_header = ['year', 'month', 'day', 'hour','product' ,'quantity', 'amount_in_gbp']
```

```python
# Function to scrape filename, alter to remove dots, underscore to eventually form column name, tested with one filename. 
def get_storename(filename_with_ext):
    filepaths = filename_with_ext.split('.')
    filename = filepaths[-2]
    filename = filename[10: ]
    result = filename.replace('_', ' ')
    return result

get_storename('2012-2020_Shepway_store.csv')
```

```python
#Iteration through list in directory and creation of new formatted header using list of pre-defined column header titles called new header

os.chdir('C:\\Users\\Public\Documents\\MLW-Code\\Dash-Project') # print(os.getcwd()) used to check you are in the right directory, note \\ needed to not be confused for escape command

csv_dfs = [] # create [] to store the end result
for f in os.listdir(): 
    if f.endswith('csv'):
        dataframe_with_new_header = pd.read_csv(f, header=0, names=new_header) #This adds a header with consistent, specified titles not the existing header
        dataframe_with_new_header['branch_name']=get_storename(f) #This adds branch_name to the data and runs function to alter existing filename insert into this column
        csv_dfs.append(dataframe_with_new_header) # This adds the above data into the empty Python List called csv_dfs
        # print(f)
        # s = get_storename(f)
        # print(csv_dfs) # allows initial files to be tested to debug 
        # print(s, ' \n\n')
all_csv_files = pd.concat(csv_dfs) # the end concatenation of all csv files  
all_csv_files




```

```python
# used regex to remove non-numeric data and keep look of existing data with inplace=True, quantity then transformed into numeric data 

all_csv_files['quantity'].replace(to_replace='[^0-9]+', value='',inplace=True,regex=True)
all_csv_files["quantity"] = pd.to_numeric(all_csv_files["quantity"], errors='ignore')
all_csv_files
```

```python
# the same required for the amount_in_gbp column 
# producing results such as # all_csv_files['amount_in_gbp'].replace('23.04000000d0000006', 0, inplace=True)
# all_csv_files['amount_in_gbp'].replace('5.279c999999999999', 0, inplace=True)
# all_csv_files['amount_in_gbp'].replace('382c52.8', 0, inplace=True)

all_csv_files['amount_in_gbp'].replace(to_replace='[^0-9]+', value=0 ,inplace=True,regex=True)
all_csv_files["amount_in_gbp"] = pd.to_numeric(all_csv_files["amount_in_gbp"], errors='ignore')
```

```python
# As initially the amount_in_gbp was converted to be a float64 astype used to convert with numpy. Cannot convert non-finite values (NA or inf) hence used fillna to fix 
all_csv_files['quantity'] = all_csv_files['quantity'].fillna(0).astype(np.int64)
all_csv_files.dtypes

```

```python
#iterating through each file in the directory ending in json to add the new header and use function get_storename to use part of filename as column branch_name

os.chdir('C:\\Users\\Public\Documents\\MLW-Code\\Dash-Project') # print(os.getcwd()) used to check you are in the right directory, note \\ needed to not be confused for escape command

json_dfs = [] # create [] to store the end result
for f in os.listdir():   #iterating through f for file within the directory list for all the files ending with json
    if f.endswith('json'):
        dataframe_with_new_header = pd.read_json(f) #This adds a header with consistent, specified titles not the existing header
        dataframe_with_new_header.set_axis(new_header, axis=1, inplace=True)
        dataframe_with_new_header['branch_name']=get_storename(f) #This adds branch_name to the data and runs function to alter existing filename insert into this column
        json_dfs.append(dataframe_with_new_header) # This adds the above data into the empty Python List called csv_dfs
        print(json_dfs) # allows initial files to be tested to debug 

all_json_files = pd.concat(json_dfs)
all_json_files
```

```python
# Similar to csv action to be able to remove non-numeric data to allow change in datatype from object to numeric values 
all_json_files['quantity'].replace(to_replace='[^0-9]+', value='',inplace=True,regex=True)
all_json_files["quantity"] = pd.to_numeric(all_json_files["quantity"], errors='ignore')
all_json_files
```

```python
# Change from float to int64 for quantity
all_json_files['quantity'] = all_json_files['quantity'].fillna(0).astype(np.int64)
```

```python
# Concatenation of CSV and Json branch data to form all branches irrespective of data extension type
combined_csv_json = [all_json_files, all_csv_files]
all_branches = pd.concat(combined_csv_json)
all_branches 
```

```python
# merge of all branch data with the products list which contains product categories 
all_branches_products = all_branches.merge(productslist_df, how='left')
all_branches_products
```

```python
# replacing fruits & vegetable to fruits & vegetables 

all_branches_products['category'].replace('fruits & vegetable', 'fruits & vegetables', inplace=True)
all_branches_products
```

```python
#A left merge with region information and branch expenses 
# region_expenses = region_df.merge(branch_expenses_df, how='left')
# region_expenses 
```

```python
# merge of all branches, region, product and product cat (no expenses)
all_branch_products_region= all_branches_products.merge(region_df,  how='left')
all_branch_products_region
```

```python
# Access to the summed amount in gbp and quantity for the region 
total_quant_gbp_per_region = all_branch_products_region.groupby(['region' , 'branch_name'])['amount_in_gbp', 'quantity'].sum().reset_index()
total_quant_gbp_per_region
```

```python
# Addition of performance column for the region 
total_quant_gbp_per_region['performance'] = total_quant_gbp_per_region['amount_in_gbp'] + total_quant_gbp_per_region['quantity']
total_quant_gbp_per_region.sort_values(by=['performance'], ascending=False, inplace=True)
total_quant_gbp_per_region
```

```python
#For brief. no.3 and no.2 this provides the performance for Region top 10 branches needed for the sales per hour display 
total_quant_gbp_per_region.head(10)
```

```python
#Next the access to the top ten performance for the County 
# The summed amount in gbp and quantity for the county, the variable all_branches_product_region also includes the county column already
total_quant_gbp_per_county = all_branch_products_region.groupby(['county' , 'branch_name'])['amount_in_gbp', 'quantity'].sum().reset_index()
total_quant_gbp_per_county
```

```python
#Addition of the performance column with addition of sales amount in gbp with quantity for the county 
total_quant_gbp_per_county['performance'] = total_quant_gbp_per_county['amount_in_gbp'] + total_quant_gbp_per_county['quantity']
total_quant_gbp_per_county.sort_values(by=['performance'], ascending=False, inplace=True)
total_quant_gbp_per_county
```

```python
#Display and test to see the top performance for the county information needed for brief no.3
total_quant_gbp_per_county.head(10)
```

```python
# For brief 3 step groupby to take branchname and date parameters of for time column for county  
all_branch_hour_sales = all_branch_products_region.groupby(['branch_name', 'year','month', 'day','hour' ])['amount_in_gbp'].sum().reset_index()
all_branch_hour_sales
```

```python
# This is a total of gbp per branch_name to ensure there are the same amount of cells for each column 
total_gbp_per_branch = all_branch_products_region.groupby('branch_name')['amount_in_gbp'].sum().reset_index()
total_gbp_per_branch
```

```python
# Totalling expenses for each branch_name using the same format 
total_expenses_per_branch = branch_expenses_df.groupby('branch_name')['total_expenses'].sum().reset_index()
total_expenses_per_branch
```

```python
# merge of gbp and expenses for the 86 rows of data in each 
total_gbp_and_expenses = total_expenses_per_branch.merge(total_gbp_per_branch, how='left')
total_gbp_and_expenses
```

```python
# For the 4th brief this shows the profitability totals and this sorted by this column, inplace retains the information 
total_gbp_and_expenses['profitability'] = total_gbp_and_expenses['amount_in_gbp'] - total_gbp_and_expenses['total_expenses']
total_gbp_and_expenses.sort_values(by=['profitability'], ascending=False, inplace=True)

```
